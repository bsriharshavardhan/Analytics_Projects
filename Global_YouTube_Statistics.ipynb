## Data cleaning

### Subtask:
Clean the loaded YouTube data.


```python
# Handling missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Fill missing values in 'subscribers_for_last_30_days' with 0, as it represents a count.
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows with missing values in 'created_month' and 'created_date' as imputation isn't suitable for categorical data.
df.dropna(subset=['created_month', 'created_date'], inplace=True)

# Fill missing values in numerical features with the mean.
numerical_cols = ['subscribers', 'video views', 'uploads', 'video_views_for_the_last_30_days', 'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings', 'highest_yearly_earnings', 'Gross tertiary education enrollment (%)', 'Population', 'Unemployment rate', 'Urban_population', 'Latitude', 'Longitude']
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())


print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Removing duplicate rows
duplicate_rows = df.duplicated()
num_duplicates = duplicate_rows.sum()
print(f"\nNumber of duplicate rows: {num_duplicates}")
df.drop_duplicates(inplace=True)

# Correcting data types
print("\nData types before correction:")
print(df.dtypes)

# Convert relevant columns to numeric types
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert 'created_year' to integer
df['created_year'] = df['created_year'].astype('Int64')

# Convert 'created_date' to integer
df['created_date'] = df['created_date'].astype('Int64')


print("\nData types after correction:")
print(df.dtypes)
display(df.head())
```

## Data exploration

### Subtask:
Examine the shape of the DataFrame, summarize descriptive statistics of numerical features, analyze the distribution of categorical features, investigate correlations, check for inconsistencies, and document observations.

```python
# Shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Descriptive statistics of numerical features
numerical_features = ['subscribers', 'video views', 'uploads', 'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings', 'highest_yearly_earnings']
print("\nDescriptive statistics of numerical features:")
display(df[numerical_features].describe())

# Distribution of categorical features
categorical_features = ['category', 'Country of origin']
for col in categorical_features:
    print(f"\nDistribution of '{col}':")
    display(df[col].value_counts())

# Correlation between numerical variables
correlation_matrix = df[numerical_features].corr()
print("\nCorrelation matrix of numerical features:")
display(correlation_matrix)

# Check for inconsistencies (e.g., negative values in count/earnings columns)
for col in ['subscribers', 'video views', 'uploads', 'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings', 'highest_yearly_earnings', 'subscribers_for_last_30_days']:
    if (df[col] < 0).any():
        print(f"\nWarning: Negative values found in '{col}' column.")

# Document observations and potential outliers
print("\nObservations and potential outliers:")
# Example: Identify potential outliers in 'subscribers'
potential_subscriber_outliers = df[df['subscribers'] > df['subscribers'].quantile(0.95)]
print("Potential outliers in 'subscribers':")
display(potential_subscriber_outliers[['Youtuber', 'subscribers']])
```

## Data analysis

### Subtask:
Analyze the YouTube data to answer specific questions.


```python
# 1. Top 10 Channels by Subscribers
top_10_channels = df.sort_values(by='subscribers', ascending=False).head(10)
print("Top 10 YouTube channels by subscribers:\n", top_10_channels[['Youtuber', 'subscribers']])

# 2. Category with Highest Average Subscribers
category_avg_subscribers = df.groupby('category')['subscribers'].mean()
highest_avg_subscribers_category = category_avg_subscribers.idxmax()
print(f"\nCategory with the highest average subscribers: {highest_avg_subscribers_category}")

# 3. Average Video Uploads per Category
avg_uploads_per_category = df.groupby('category')['uploads'].mean()
print("\nAverage video uploads per category:\n", avg_uploads_per_category)

# 4. Top 5 Countries with Most Channels
top_5_countries = df['Country of origin'].value_counts().head(5)
print("\nTop 5 countries with the most YouTube channels:\n", top_5_countries)

# 5. Distribution of Channel Types across Categories
if 'channel_type' in df.columns:
    channel_type_distribution = df.groupby(['category', 'channel_type']).size().unstack(fill_value=0)
    print("\nDistribution of channel types across categories:\n", channel_type_distribution)
else:
    print("\n'channel_type' column not found. Using 'category' for distribution.")
    category_distribution = df['category'].value_counts()
    print("\nDistribution of categories:\n", category_distribution)

# 6. Correlation between Subscribers and Video Views
correlation = df['subscribers'].corr(df['video views'])
print(f"\nCorrelation between subscribers and video views: {correlation}")

# 7. Monthly Earnings Variation across Categories
# Using both lowest and highest monthly earnings for a more comprehensive analysis
avg_lowest_earnings = df.groupby('category')['lowest_monthly_earnings'].mean()
avg_highest_earnings = df.groupby('category')['highest_monthly_earnings'].mean()
print("\nAverage monthly earnings variation across categories:")
print("Lowest Monthly Earnings:\n", avg_lowest_earnings)
print("\nHighest Monthly Earnings:\n", avg_highest_earnings)
```

```python
# 8. Trend in Subscriber Growth over the Last 30 Days
total_subscriber_growth = df['subscribers_for_last_30_days'].sum()
print(f"\nTotal subscriber growth in the last 30 days: {total_subscriber_growth}")

# 9. Outliers in Yearly Earnings
# Calculate the IQR for 'highest_yearly_earnings'
Q1 = df['highest_yearly_earnings'].quantile(0.25)
Q3 = df['highest_yearly_earnings'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['highest_yearly_earnings'] < lower_bound) | (df['highest_yearly_earnings'] > upper_bound)]
print("\nOutliers in yearly earnings:\n", outliers[['Youtuber', 'highest_yearly_earnings']])
```
